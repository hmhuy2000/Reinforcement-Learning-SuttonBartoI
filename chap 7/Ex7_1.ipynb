```python
# ========================= import ===================================

import numpy as np 
import matplotlib.pyplot as plt
from tqdm import tqdm, trange
import seaborn as sns
import random
```

```python


# ======================== CFG ======================================

class CFG:

    episode = 10
    TD_alpha = np.arange(0, 1.1, 0.1)
    target = np.arange(-20, 22, 2) / 20.0
    run = 100
    garma = 1.0
```

```python


# ======================== init ======================================

class game():
    def __init__(self, alpha, TypeAlgo, n_step):
        self.value_function = np.full((21), 0.0)
        self.policy = np.full((21,2), 0.5)
        self.Left = -1
        self.Right = 1
        self.Start = 10
        self.Stop = 0
        self.n_step = n_step
        self.actions = [self.Left, self.Right]
        self.FirstEnd = 0
        self.LastEnd = 20
        self.alpha = alpha
        self.TypeAlgo = TypeAlgo

    def reset(self):
        pass

    def step(self):
        if (self.TypeAlgo == 'TD'):
            return self.step_TD()

    def step_TD(self):
        his = []
        cur = self.Start
        reward = 0
        T = float('inf')
        time = 0
        done = 0
        while(time - self.n_step  < T - 1):
            if ((cur == self.LastEnd or cur == self.FirstEnd) and done == 0):
                if cur == self.LastEnd:
                    state_reward = 1
                elif cur  == self.FirstEnd:
                    state_reward = -1
#                 print((cur , self.Stop, state_reward))
                his.append((cur , self.Stop, state_reward))
                T = time 
                done = 1
            
            
            if (time < T):
                action = np.random.choice(self.actions, p=self.policy[cur, :])
                state_reward = 0
#                 print((cur, action, state_reward))
                his.append((cur, action, state_reward))
            
            update_time = time - self.n_step + 1
            if (update_time >= 0 and update_time<len(his)):
                G = 0.0
                for t in range(update_time + 1, min(T + 1, update_time + self.n_step)+1):
                    (_,_,S_reward) = his[t - 1]
                    G += pow(CFG.garma, t - update_time -1) * S_reward
                if (update_time + self.n_step < T):
                    G += pow(CFG.garma,self.n_step) * self.value_function[cur + action]
#                 print(G, his[min(T + 1, update_time + self.n_step)-1],his[-1])
                (update_state, _, _) = his[update_time]
                self.value_function[update_state] += self.alpha\
                * (G - self.value_function[update_state])
                
            if(done == 0):
                cur += action
                reward += state_reward
            time += 1

        return his, reward

    def get_value_function(self):
        return self.value_function
```

```python


# ======================= function ===================================

def TD_evaluate(episode, alpha, n_step):
    total_error = 0.0
    t_tqdm = trange(CFG.run, desc = f'TD n = {n_step} alpha = {alpha:.2f}')
    for r in t_tqdm:
        env = game(alpha = alpha, TypeAlgo = 'TD', n_step = n_step)
        error = []
        for ep in range(episode):
            env.reset()
            his, reward = env.step()
#             print(env.get_value_function())
            total_error += (np.sqrt(np.sum(np.power(env.get_value_function()[1:-1] - CFG.target[1:-1], 2))/19.0))
        t_tqdm.set_description(f'TD n = {n_step} a = {alpha:.2f} = {total_error/((r+1) * CFG.episode):.3f}')
    return total_error/(CFG.run*CFG.episode)
```

```python

# ======================= main ========================================
res = np.zeros((10, 11))
for x in range(10):
    n_step = pow(2,x)
    for idx in range(len(CFG.TD_alpha)):
        alpha = CFG.TD_alpha[idx]
        error = TD_evaluate(CFG.episode, alpha, n_step)
        res[x,idx] = error
        
#         print(f'{n_step} - {alpha} = {error}')

```

```python
from matplotlib.pyplot import figure
for x in range(10):
    n_step = pow(2,x)
    plt.plot(CFG.TD_alpha,res[x,:11], label = f'n = {n_step}')
plt.xlabel('alpha')
plt.ylabel('RMS error')
plt.rcParams["figure.figsize"] = (10,3)
plt.ylim([0.2, 0.8])
plt.legend()
plt.savefig('figure_Ex7_1/figure_Ex7_1.png')
plt.show()
```
